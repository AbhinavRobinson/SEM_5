## print dimesions of frame 4

```python
cap = cv2.VideoCapture('/Users/regatte/Desktop/challenge-week-1/data/videos/VID_6.mp4')
count=0

while(cap.isOpened()):
    ret, frame = cap.read()
    if count==0:
        width  = cap.get(cv2.CAP_PROP_FRAME_WIDTH)   
        height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
        print('width, height:', width, height)
        count+=1

    
    if frame is None:
        break
    cv2.imshow('app', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
cap.release()
cv2.destroyAllWindows()
```

## pixel values

```python
cap = cv2.VideoCapture('/Users/regatte/Desktop/challenge-week-1/data/videos/VID_6.mp4')
count=0

while(cap.isOpened()):
    ret, frame = cap.read()
    if count==10:
        print(frame.shape, type(frame))
        print("dimensions- {},{}".format(frame.shape[0], frame.shape[0]))
        print("pixel values -{}".format(frame))
        count+=1
    else:
        count+=1


    
    if frame is None:
        break
    cv2.imshow('app', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
cap.release()
cv2.destroyAllWindows()
```

## check dimesions equal

```python
for i in range(1,15):    
    cap = cv2.VideoCapture('/Users/regatte/Desktop/challenge-week-1/data/videos/VID_14.mp4'.format(i))
    count=0

    while(cap.isOpened()):
        ret, frame = cap.read()
        
        if frame is None:
            break
        
        if count==0:
            h,w,c=(frame.shape[0], frame.shape[1], frame.shape[2])
            count+=1
           
        else:

            if(h!=frame.shape[0] or w!= frame.shape[1] or c!=frame.shape[2]):
                print("video {}, dimesions unequal".format(i) )
            count+=1
            

        cv2.imshow('app', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        

        
        
    cap.release()
    cv2.destroyAllWindows()
```
